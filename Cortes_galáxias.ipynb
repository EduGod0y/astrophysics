```python
#@title NGC 4244
# Basicamente o mais importante está aqui
!pwd

import matplotlib.pyplot as plt
from astropy.table import Table
data = Table.read('NGC2403-PM')  # Aqui carrega a tabela com todos dados
data2 = Table.read('NGC2403-RA,DEC,MAG,GAIA')
#mag_g = data['MAG_AUTO_g']-0.150
#mag_r = data['MAG_AUTO_r']-0.101
#colour= data['g-r']
#star = data['CLASS_STAR_r']
#fwhm = data['FWHM_IMAGE_2']
#summ =0

#data2 = Table.read('cut_06.fits')     # Aqui eu abro a tabela com os cortes de match que fiz no topcat

# Aqui eu faço outros cortes usando parametros das tabelas
subset = data[((data['mag_g']-data['mag_i'])<=2)  & ((data['mag_r'])<=20) & ((data['mag_g']-data['mag_i'])>=0) & ((data['mag_g'])<35)  & ((data['mag_i'])<35)] 
subset2 = data2[((data2['mag_g']-data2['mag_i'])<=2)  & ((data2['mag_r'])<=20) & ((data2['mag_g']-data2['mag_i'])>=0) & ((data2['mag_g'])<35)  & ((data2['mag_i'])<35)] 
#print(subset)

# Para plotar

#fig = plt.figure(figsize=(12,12))

#plt.rc('legend', fontsize=19) 
#plt.rc('font', size=28) 
plt.scatter(subset2['mag_g']-subset2['mag_i'],subset2['mag_r'])
plt.scatter(subset['mag_g']-subset['mag_i'],subset['mag_r'])
#plt.errorbar(colour, mag_g, yerr=data['MAGERR_AUTO_r'], xerr=data['MAGERR_AUTO_r'], label='Detected objects',marker='^',color='gray',ms=11,lw=0,alpha=0.5)
plt.gca().invert_yaxis()
plt.legend()
plt.title('NGC 2403', size=25)
plt.ylabel('r [mag]', size=32)
plt.xlabel('g-i [mag]', size=32)
plt.show()
```
