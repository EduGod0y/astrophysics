```python
#@title Função redução dos objetos
# mag_j-plus, mag_sextractor, class star, fwhm, cut mag_j-plus, cut class star, cut fwhm, j-plus error, sextractor error
def zero(mag, mag_sex, cs, fwhm, err_jplus, err_sex, x1, x2, x3):
  cond = (mag < x1) & (cs > x2) & (fwhm < x3)

  selected_mag = mag[cond]
  selected_mag_sex = mag_sex[cond]
  selected_err_jplus = err_jplus[cond]
  selected_err_sex = err_sex[cond]

  zero_point = np.mean(np.array(selected_mag_sex-selected_mag))
  var_zero_point = np.var(np.array(selected_mag_sex-selected_mag))

  return zero_point, var_zero_point, selected_mag, selected_mag_sex, selected_err_jplus, selected_err_sex
```

```python
#@title Função gráfico
# jplus_mag, sextractor_mag, selected mag j-plus, selected mag sextractor, error j-plus, error sextractor, selected error j-plus, selected error sextractor, zero point, variation zero point, xlabel, ylabel, plot title
def graph(jplus, sex, sel_jplus, sel_sex, err_jplus, err_sex, sel_err_jplus, sel_err_sex, zp, vpz, xlabel, ylabel, title):
  x = np.arange(0,30,1)
  figura = plt.figure(figsize=(15, 10))

  plt.subplot(221)
  plt.scatter(jplus, sex, s = 5)
  plt.errorbar(jplus, sex, alpha=0.5, yerr = abs(err_sex), fmt='.', ecolor = 'k')
  plt.plot(x-(zp),x, color = 'red')
  plt.plot(x-(zp)+vpz,x,color = 'r', linestyle='dashed')
  plt.plot(x-(zp)-vpz,x,color = 'r', linestyle='dashed')
  plt.ylabel(ylabel)
  plt.xlabel(xlabel)
  plt.xlim([12,18])
  plt.ylim([12,18])

  plt.subplot(222)
  plt.scatter(sel_jplus, sel_sex-(zp), s = 5)
  plt.errorbar(sel_jplus, sel_sex-(zp), alpha=0.5, yerr = abs(sel_err_sex), fmt='.', ecolor = 'k')
  plt.plot(x,x, color = 'red')
  plt.plot(x+vpz,x,color = 'r', linestyle='dashed')
  plt.plot(x-vpz,x,color = 'r', linestyle='dashed')
  plt.ylabel(ylabel)
  plt.xlabel(xlabel)
  plt.xlim([12,18])
  plt.ylim([12,19])

  plt.subplot(223)
  plt.scatter(jplus, sex-jplus, s = 5)
  plt.errorbar(jplus, sex-jplus, alpha=0.5, yerr = abs(err_sex-err_jplus), fmt='.', ecolor = 'k')
  plt.axhline(y=zp, color = 'r')
  plt.axhline(y=zp+vpz, color = 'r', linestyle='dashed')
  plt.axhline(y=zp+-vpz, color = 'r', linestyle='dashed')
  plt.ylabel('{} - {}'.format(ylabel,xlabel))
  plt.xlabel(xlabel)
  plt.xlim([12,18])
  plt.ylim([-0.2,0.2])

  plt.subplot(224)
  plt.scatter(sel_jplus, sel_sex-(zp)-sel_jplus, s = 5)
  plt.errorbar(sel_jplus, sel_sex-(zp)-sel_jplus, alpha=0.5, yerr = abs(sel_err_sex-sel_err_jplus), fmt='.', ecolor = 'k')
  plt.axhline(y=0, color = 'r')
  plt.axhline(y=vpz, color = 'r', linestyle='dashed')
  plt.axhline(y=-vpz, color = 'r', linestyle='dashed')
  plt.ylabel('{} - {}'.format(ylabel,xlabel))
  plt.xlabel(xlabel)
  plt.xlim([12,18])
  plt.ylim([-0.2,0.2])

  plt.suptitle('Calibration of NGC 2403, with the {}-band'.format(title))

  plt.show()
  return figura
```

```python
#@title Seleção de GCs
# mag_j-plus, mag_sextractor, class star, fwhm, cut mag_j-plus, cut class star, cut fwhm, j-plus error, sextractor error
def plot(r, g, i, u, z, ra, dec, cs_r, cs_g, cs_i, pm):
  cond = (r < 20) & (g-i < 2) & (0 < g-i) & (u-i < 2) & (-2 < u-i) & (i-z < 6) & (-2 < i-z) & (g-z < 5) & (z-i < 2) & (-1 < g-z) & (-2 < z-i) & (pm < 2.5)
 
  cond2 = (r < 15) & (g-i < 2) & (0 < g-i) & (u-i < 2) & (-2 < u-i) & (i-z < 6) & (-2 < i-z) & (g-z < 5) & (z-i < 2) & (-1 < g-z) & (-2 < z-i)

  selected_r = r[cond]
  selected_g = g[cond]
  selected_i = i[cond]
  selected_u = u[cond]
  selected_z = z[cond]
  selected_ra = ra[cond]
  selected_dec = dec[cond]
  selected_pm = pm[cond]
  selected_cs_r = cs_r[cond]


  selected_r_2 = r[cond2]
  selected_g_2 = g[cond2]
  selected_i_2 = i[cond2]
  selected_u_2 = u[cond2]
  selected_z_2 = z[cond2]
  selected_ra_2 = ra[cond2]
  selected_dec_2 = dec[cond2]
  selected_pm_2 = pm[cond2]

  return selected_r, selected_g, selected_i, selected_u, selected_z, selected_ra, selected_dec, selected_r_2, selected_g_2, selected_i_2, selected_u_2, selected_z_2, selected_ra_2, selected_dec_2, selected_pm, selected_pm_2, selected_cs_r
```
